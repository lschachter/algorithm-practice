```python
class Permutations:
    '''
    Given a collection of distinct integers, return all possible permutations.
    
    This one tripped me up for an incredibly long time. 
    '''
    def permute(self, nums: List[int]) -> List[List[int]]:
        self.nums = nums
        self.perms = []
        self.n = len(nums)
        self.permuteHelper(0)
        return self.perms
        
        
    def permuteHelper(self, i):
        if i == self.n - 1:
            self.perms.append(list(self.nums))
            return
        for j in range(i, self.n):
            self.nums[i], self.nums[j] = self.nums[j], self.nums[i]
            self.permuteHelper(i + 1)
            self.nums[i], self.nums[j] = self.nums[j], self.nums[i]
        return
        
```

```python
class CountUniquePaths:
    '''
    A robot is located at the top-left corner of a m x n grid.
    The robot can only move either down or right at any point in time. The robot is trying to 
    reach the bottom-right corner of the grid.
    How many possible unique paths are there?
    
    
    '''
    def uniquePaths(self, m: int, n: int) -> int:
        self.n = n
        self.m = m
        self.counts = {}
        return self.countPaths(0,0)
        
    def countPaths(self, row, col):
        if (row, col) in self.counts: return self.counts[(row, col)]
        if row >= self.n or col >= self.m: return 0
        if row == self.n - 1 and col == self.m - 1: return 1
        numR = self.countPaths(row, col + 1)
        numL = self.countPaths(row + 1, col)
        self.counts[(row, col)] = numR + numL
        return numR + numL
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class ClosestBSTVal:
    '''
    Given a non-empty binary search tree and a target value, find the value in the BST that is 
    closest to the target.
    Note: Given target value is a floating point
        You are guaranteed to have only one unique value in the BST that is closest to the target
          
          
    We don't need to check every node in the tree, we only need to look at the ones effectively on
    either side of the target. so compare the target to the current node's value, track it if it's
    smaller than your minDiff (don't go any further down the tree if the target == node.val), then
    decide which way to go based on whether the target is smaller or larger than the current
    node's value.
    target = 3.7
            4         # minDiff = .3, minVal = 4, 3.7 < 4 so call again with 4.left = 2
          2   5       # currDiff = abs(2 - 3.7) = 1.7, not < minDiff so ignore and call again. 
                        3.7 > 2 so go right
        1  3          # currDiff = abs(3 - 3.7) = .7 > minDiff so ignore; call again with 3.right
                      # 3.right = None so you return and recurse through the tree
        
    we are following a single path down the tree, so our runtime is O(h) where h = height of the
    tree. Worst case, the tree only has a single branch for us to follow in which case the runtime
    is O(n). space-wise, we aren't using any auxilary data structures so it's just the space of
    the call stack which is again O(h).
    '''
    def closestValue(self, root: TreeNode, target: float) -> int:
        self.t = target
        self.minVal = root.val
        self.minDiff = abs(root.val - self.t)
        self.closestHelper(root)
        return self.minVal
    
    def closestHelper(self, node):
        if not node: return
        currDiff = abs(node.val - self.t)
        if currDiff < self.minDiff:
            self.minVal = node.val
            self.minDiff = currDiff
        if node.val == self.t: return
        if self.t < node.val:
            self.closestHelper(node.left)
        else:
            self.closestHelper(node.right)
```

```python
class FirstLastPositionSorted:
    '''
    Given an array of integers nums sorted in ascending order, find the starting and ending position
    of a given target value.
    Your algorithm's runtime complexity must be in the order of O(log n).
    If the target is not found in the array, return [-1, -1].
    
    Since the array is sorted, we can use binary search to find the target. Once we have found the target,
    we can do another binary search on either side of the target for the next number, which is definitely
    ideal if we're expecting large ranges of the target number. if we aren't though, we can also do a linear
    2 pointer search going out to either side from the element we've found that == the target until we hit
    a different number on either side. this is worst case O(n) when the entire array is made up of the target
    value but listen. I'm tryina do these timed and i ain't got the time for that. 
    
    We can't break the array down into smaller bits without having to do some funky addition through the
    recursion to keep track of where the indices from the smaller chunk should be added/subtracted from mid
    etc, so it seemed easier to just track the start and end of where in the array we're looking at each
    recursion level. once start >= end, we know we haven't found the target, so return -1, -1. if we do find
    it, do the two pointer search for the ends of the range. the space is O(logn) for the call stack.
    '''
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        self.nums = nums
        return self.searchRangeHelper(target, 0, len(self.nums))
        
    def searchRangeHelper(self, target, start, end):
        n = end - start + 1
        if start >= end: return [-1,-1]
        mid = start + (end - start) // 2
        if self.nums[mid] == target:
            return self.findRange(target, mid, start, end)
        if self.nums[mid] < target:
            return self.searchRangeHelper(target, mid + 1, end)
        return self.searchRangeHelper(target, start, mid)
    
    def findRange(self, target, mid, start, end):
        p1 = p2 = mid
        while p1 - 1 >= start and self.nums[p1 - 1] == target:
            p1 -= 1
        while p2 + 1 < end and self.nums[p2 + 1] == target:
            p2 += 1
        return [p1, p2]
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class BSTInsertion:
    '''
    Given the root node of a binary search tree (BST) and a value to be inserted into the tree, 
    insert the value into the BST. Return the root node of the BST after the insertion. 
    It is guaranteed that the new value does not exist in the original BST.
    Note that there may exist multiple valid ways for the insertion, as long as the tree remains 
    a BST after insertion. You can return any of them.
    
    I think it's generally easier t always insert as a leaf. we don't need to worry about balancing 
    the tree so the time complexity issues that this can cause can be ignored for this problem. So,
    if no root, return the val as the root. otherwise, compare val to node's val:
    if val > node.val: if not node.right: node.right = Node(val); else: recurse with node.right
    else: if not node.left: node.left = Node(val); else: recurse with node.left
    val = 5
            4            # 5 > 4; 4.right != None so recurse with 4.right
          2   7          # 5 < 7; 7.left == None so place 5 here
        1  3
        
    We are essentially doing a depth first search for the correct placement of the new node. In the 
    worst case, where elements are inserted in order (1,2,3,4,5, for example), this creates a O(n)
    runtime for insertion, search, and deletion. which is obviously, very bad. We want O(logn) for 
    binary search trees in general, other than traversing through each node. So i suppose in an 
    interview setting, this would be considered the brute force answer, with an element of balancing
    being necessary for the best, worst case scenario. but leetcode passed it, so shrug emoji. 
    
    Space complexity is the same as time, based on the call stack.
    '''
    def insertIntoBST(self, root: TreeNode, val: int) -> TreeNode:
        if not root:
            return TreeNode(val)
        self.val = val
        self.insertNode(root)
        return root
        
    def insertNode(self, node):
        if self.val > node.val: # move right
            if not node.right:
                node.right = TreeNode(self.val)
                return
            self.insertNode(node.right)
        else:
            if not node.left:
                node.left = TreeNode(self.val)
                return
            self.insertNode(node.left)
```

```python
class Fibonacci:
    '''
    Fibonacci Numbers = 
    F(0) = 0,   F(1) = 1
    F(N) = F(N - 1) + F(N - 2), for N > 1.
    for all positive integers.
    
    So the above is exactly the base case and recursive case for the recursive solution to this problem.
    adding a guard against negative numbers, that's truly the code. I separated the guard into a runner
    program so that it didn't need to check if the input was negative at every recursion level:
    
    def fib(self, N: int) -> int:
        if N < 0: return -1
        return self.doFib(N)
        
    def doFib(self, N):
        if N == 0 or N == 1: return N
        return self.doFib(N -1) + self.doFib(N - 2)
        
    but in this case, an iterative solution is actually significantly faster-- in this simple recursive
    answer, fib(n - 1) and fib(n - 2) both need to be calculated down to 1, even though all but one of 
    those calls is running the exact same calculation (n = 10 so fib(9) + fib(8), you have to calculate
    fib(8) twice, which breaks down into fib(8) + fib(7) and again and again from there. basically at each
    recursive call we make 2 more recursive calls. this results in 2^n runtime. whereas iteration allows
    you to only calculate each number's answer once, saving an incredible amount of time and call stack 
    space. Alternatively, you could recurse and keep track of a dictionary of answers to each number as
    you calculate it, then check the dictionary before recursing. But iteration is still better in this 
    case, because space complexity is O(n) in that recursive solution with time complexity also O(n), where
    the iterative approach is O(n) time and O(1) space. 
    '''
    def fib(self, N: int) -> int:
        if N < 0: return -1
        if N == 0 or N == 1: return N
        
        a = 0
        b = 1
        c = a + b
        for i in range(N - 2):
            a = b
            b = c
            c = a + b
        return c
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class SameTree:
    '''
    Given two binary trees, write a function to check if they are the same or not.
    Two binary trees are considered the same if they are structurally identical and the nodes 
    have the same value.
    
    So you can do a general traversal of the trees and at each node, check if the both don't 
    exist (in which case you can just return, they're the same so you don't want to change
    the self.same flag). then you can check if one or the other doesn't exist or if their
    values are not the same, in which case, change the flag to false and return. Other than
    these two base cases, continue the traversal with the left and right nodes. 
    '''
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        self.same = True
        self.isSameNode(p, q)
        return self.same
    
    def isSameNode(self, node1, node2):
        if not node1 and not node2: return
        if not node1 or not node2 or node1.val != node2.val:
            self.same = False
            return
        self.isSameNode(node1.left, node2.left)
        self.isSameNode(node1.right, node2.right)
```

```python
class findRotatedMin:
    '''
    Suppose an array sorted in ascending order is rotated at some pivot unknown to you beforehand.
    (i.e., [0,1,2,4,5,6,7] might become [4,5,6,7,0,1,2]).
    Find the minimum element.
    You may assume no duplicate exists in the array.
    
    Ok so this will be a binary search, except instead of knowing the target value, we're looking
    for an element less than the current mid to the right of it. within that, keep track of the min
    you've found, because in a situation like [7,8,0,1,2] 0 < 2 so you'll recurse left, but 0 is the
    smallest. this brings up another point-- if self.nums[mid] > self.minVal, you can stop the
    recursion because everything to the left is going to be bigger than the min.
    [4,5,6,7,0,1,2]    # 7 > 2 so check right
     s     m     e
     
    [4,5,6,7,0,1,2]     # 0 < 2 set min to 0 then recurse left in case min is further left
           s m   e
           
    runtime is O(logn) because of the binary search, and space complexity is O(logn) because
    of the call stack
    '''
    def findMin(self, nums: List[int]) -> int:
        import math
        self.minVal = math.inf
        self.nums = nums
        self.findLocalMin(0, len(nums) - 1)
        if self.minVal == math.inf: return
        return self.minVal
    
    def findLocalMin(self, start, end):
        if start > end: return
        if start == end:
            self.minVal = min(self.minVal, self.nums[start])
            return
        mid = start + (end - start) // 2
        if self.nums[mid] <= self.nums[end]:
            if self.minVal < self.nums[mid]:
                return
            self.minVal = self.nums[mid]
            self.findLocalMin(start, mid)
        else:
            self.findLocalMin(mid + 1, end)
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class SumOfLeftLeaves:
    '''
    Find the sum of all left leaves in a given binary tree.
        3
       / \
      9  20
        /  \
       15   7

    There are two left leaves in the binary tree, with values 9 and 15 respectively. Return 24.
    
    we need to hit all the leaves in the tree to make sure we've seen all the left leaves, so 
    we can start from a basic traversal, keeping track of whether or not it was a left child. 
    at each node, check if it's a leaf and if so, if it was a left child. if both are true, update
    the sum. after the traversal is over, return the sum. 
    
    we need to hit every node in the tree to make sure we've found all the left leaves, so the 
    runtime will be O(n) as will the space complexity due to the call stack.
    '''
    def sumOfLeftLeaves(self, root: TreeNode) -> int:
        self.leftSum = 0
        self.leftSumHelper(root, False)
        return self.leftSum
    
    def leftSumHelper(self, node, fromLeft):
        if node:
            self.leftSumHelper(node.left, True)
            if not node.right and not node.left and fromLeft:
                self.leftSum += node.val
            else:
                self.leftSumHelper(node.right, False)
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class InvertTree:
    '''
    Invert a binary tree.
    Example:
    Input:
         4
       /   \
      2     7
     / \   / \
    1   3 6   9
    Output:
         4
       /   \
      7     2
     / \   / \
    9   6 3   1
    
    ok so at each node you have to invert that node's left and right pointers and then continue
    in a basic traversal. pretty sure that's 100% it.
    
    again we have to traverse the tree, so runtime and space are both O(n).
    '''
    def invertTree(self, root: TreeNode) -> TreeNode:
        if root:
            root.left, root.right = root.right, root.left
            self.invertTree(root.left)
            self.invertTree(root.right)
        return root
```

```python
class SearchMatrix:
    '''
    Write an efficient algorithm that searches for a value in an m x n matrix. 
    This matrix has the following properties:
    Integers in each row are sorted from left to right.
    The first integer of each row is greater than the last integer of the previous row.
    
    ok so i think the idea is that we need to do a binary search to find which row we 
    need, and then from there a classic binary search on the row we came up with.
    I arbitrarily chose to use the last elements in each row to compare to.
    the special cases of this are that a target may be less than the element being
    compared, but it may still be in that row. therefore, when start = end, we actually
    need to check that row. 
    
    This is kind of a hilarious solution. The runtime is O(logn + logm), as is the space
    complexity. wtf hahaha. 
    '''
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        if not matrix or not matrix[0]: return False
        self.grid = matrix
        self.t = target
        self.found = False
        rowI = self.findRow(0, len(self.grid) - 1)
        if rowI == -1: return False
        if self.found: return self.found
        self.row = self.grid[rowI]
        self.findNum(0, len(self.row) - 1)
        return self.found
    
    def findRow(self, start, end):
        if start > end:
            return -1
        mid = start + (end - start) //2
        if self.grid[mid][-1] == self.t:
            self.found = True
            return
        if start == end:
            return start
        if self.grid[mid][-1] > self.t:
            return self.findRow(start, mid)
        else:
            return self.findRow(mid + 1, end)
            
    def findNum(self, start, end):
        if start >= end: return
        mid = start + (end - start) //2
        if self.row[mid] == self.t:
            self.found = True
            return
        if self.row[mid] > self.t:
            self.findNum(start, mid)
        else:
            self.findNum(mid + 1, end)
        
```
